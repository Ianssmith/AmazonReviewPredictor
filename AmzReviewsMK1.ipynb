```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Amazon.csv')
print(data.shape)
#create boolean column of repeated UserIDs
data['duplicateIds'] = data.UserId.duplicated()
data.shape
data.duplicateIds.shape
```

```python
data.head(5)
```

```python
#remove rows with duplicate UserIds
dataMD = data[data.duplicateIds <1]
dataMD.shape
#type(dataMD)
dataMD.head(5)
```

```python
idcounts = data.UserId.value_counts(sort=0)
print(len(idcounts))
idcounts.values
```

```python
#convert to numpy ndArray
idcountsM = idcounts.as_matrix()
type(idcountsM)
```

```python
#add counts column to trimmed dataframe
dataMD['ReviewCounts'] = idcountsM
dataMD[10000:10100]
dataMD.head(100)
```

```python
#transform Text feature instances to sparse matrix for hashing
from sklearn.feature_extraction.text import HashingVectorizer
hv = HashingVectorizer(n_features=2 ** 17, non_negative=True) ##use n_features= to restrict features
X = hv.transform(dataMD.Text)
#type(X)
#print(X)
```

```python
dataMD['hasSC'] = dataMD['Text'].str.contains(';')
dataMD['hasbroken'] = dataMD['Text'].str.contains('broken')
dataMD['hasdidnt'] = dataMD['Text'].str.contains('didn\'t')
dataMD['hasterrible'] = dataMD['Text'].str.contains('terrible')
dataMD['hasperfect'] = dataMD['Text'].str.contains('perfect')
dataMD['haslove'] = dataMD['Text'].str.contains('love')
dataMD['hasEP'] = dataMD['Text'].str.contains('!')
dataMD['hasdash'] = dataMD['Text'].str.contains('-')

dataMD['reviewLen'] = dataMD['Text'].str.len()
```

```python
x1 = dataMD.iloc[:, 15].values.reshape(dataMD.shape[0], 1)
x2 = dataMD.iloc[:, 16].values.reshape(dataMD.shape[0], 1)
x3 = dataMD.iloc[:, 17].values.reshape(dataMD.shape[0], 1)
x4 = dataMD.iloc[:, 18].values.reshape(dataMD.shape[0], 1)
x5 = dataMD.iloc[:, 19].values.reshape(dataMD.shape[0], 1)
x6 = dataMD.iloc[:, 20].values.reshape(dataMD.shape[0], 1)
x7 = dataMD.iloc[:, 21].values.reshape(dataMD.shape[0], 1)
x8 = dataMD.iloc[:, 22].values.reshape(dataMD.shape[0], 1)


XScore = dataMD.iloc[:, 7].values.reshape(dataMD.shape[0], 1)
XreviewLen = dataMD.iloc[:, 23].values.reshape(dataMD.shape[0], 1)

Xcounts = dataMD.iloc[:, 14].values.reshape(dataMD.shape[0], 1)

xtoadd = np.concatenate((x1, x2, x3, x4, x5, x6, x7, x8, XScore, XreviewLen, Xcounts), axis=1)
```

```python
#from sklearn.feature_extraction.text import HashingVectorizer
#hv = HashingVectorizer(n_features=2 ** 17, non_negative=True) ##use n_features= to restrict features
#X = hv.transform(dataMD.UserId)
```

```python
from scipy.sparse import csr_matrix, hstack
XtoaddSparse = csr_matrix(xtoadd)
Xfinal = hstack([X, XtoaddSparse])
X = csr_matrix(Xfinal)
X.shape
```

```python
y = dataMD.iloc[:, 12].values#.reshape(dataMD.shape[0], 1)
y.shape
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
y_pred = lr.fit(X_train, y_train).predict(X_train)
y_pred_test = lr.predict(X_test)
```

```python
def print_results():
    print('Error rate on training set: ')
    print((y_train != y_pred).sum() / X_train.shape[0])
    print('Accuracy rate on training set: ')
    print(1 - (y_train != y_pred).sum() / X_train.shape[0])
    print('True positive rate on training tet:')
    print(((y_train==True) & (y_pred==True)).sum() / y_train.sum())
    print('**************')
    print('Error rate on test set: ')
    print((y_test != y_pred_test).sum() / X_test.shape[0])
    print('Accuracy rate on test set: ')
    print(1 - (y_test != y_pred_test).sum() / X_test.shape[0])
    print('True positive rate on test set')
    print(((y_test==True) & (y_pred_test==True)).sum() / y_test.sum())
    print('True negative rate on test set')
    print(((y_test==False) & (y_pred_test==False)).sum() / (y_test.shape[0] - y_test.sum()))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler(with_mean=False)
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn import linear_model
clf = linear_model.SGDClassifier()
clf.fit(X_train_std, y_train)
y_pred = clf.fit(X_train_std, y_train).predict(X_train_std)
y_pred_test = clf.predict(X_test_std)
print_results()
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train_std, y_train)
y_pred = clf.fit(X_train_std, y_train).predict(X_train_std)
y_pred_test = clf.predict(X_test_std)
print_results()
```

```python
from sklearn import linear_model
clf = linear_model.SGDClassifier(loss='log', n_iter=50, alpha=0.00001)
clf.fit(X_train_std, y_train)
y_pred = clf.fit(X_train_std, y_train).predict(X_train_std)
y_pred_test = clf.predict(X_test_std)
print_results()
```
